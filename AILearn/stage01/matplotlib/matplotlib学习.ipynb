## 使用数据画图

```python
days = ['20-06-' + "{0:2d}".format(x) for x in range(1, 31, 1)]
temps = np.random.uniform(36, 37, size=(30))
temps2 = np.random.uniform(36, 37, size=(30))
```

```python
# 图形交互
# 声明图形
plt.figure(figsize=(16, 4))    # 设置图形长和高
# 绘制图形
plt.plot(days, temps, 'm--')
plt.plot(days, temps2, 'g.:')
plt.yticks([x for x in np.arange(35.0, 39.1, 0.2)])   # 设置坐标刻度
plt.xticks(rotation=60)
plt.title("温度-时间", fontsize=20)
plt.xlabel("time")
plt.ylabel("体温")
plt.grid(which='major', axis='both', linestyle=':', alpha=0.8)
# 显示图形
plt.show()
```
